```polyglot-notebook
[<RequireQualifiedAccess>]
type Text =
    | Number of indices: int list * number: uint
    | Symbol of index: int * symbol: char
    | Dot

let (|IsNumber|IsSymbol|IsDot|) (x: char) =
    match UInt32.TryParse(string x) with
    | true, _ -> IsNumber
    | false, _ ->
        match x with
        | '.' -> IsDot
        | _ -> IsSymbol
```

```polyglot-notebook
let parseText (line: string) =
    List.ofSeq line
    |> List.mapi (fun index character ->
        match character with
        | IsNumber -> Text.Number (index |> List.singleton, UInt32.Parse(string character))
        | IsSymbol -> Text.Symbol (index, character)
        | IsDot -> Text.Dot
    )
```

```polyglot-notebook
let rec aggregateNumbersInText (texts: Text list) =
    match texts with
    | first :: second :: tail -> 
        match first, second with
        | Text.Number (indices1, num1), Text.Number (indices2, num2) ->
            let newNumber = string num1 + string num2 |> uint
            Text.Number (indices1 @ indices2, newNumber) :: tail |> aggregateNumbersInText
        | Text.Number (indices1, num1), _ -> [Text.Number (indices1, num1)]
        | _ -> []
    | head :: tail -> 
        match head with
        | Text.Number (indices1, num1) -> [Text.Number (indices1, num1)]
        | _ -> aggregateNumbersInText tail
    | [] -> []
```
